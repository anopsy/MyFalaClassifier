```python
VID_FILES = '/content/sample_data/VID_FILES/warun/'
FRAMES = '/content/gdrive/MyDrive/FRAMES'
```

```python
unziped = tarfile.open('/content/gdrive/MyDrive/warun.tar.gz')
unziped.extractall(VID_FILES)
unziped.close()
```

```python

count = 0

for file in sorted(os.listdir(VID_FILES)):
  if file.endswith('ts'):
    filePath = os.path.join(VID_FILES, file)
    mp4FilePath = os.path.join(VID_FILES, os.path.splitext(file)[0] + ".mp4")
    if os.path.isfile(mp4FilePath):
      continue
    os.system("ffmpeg -i " + "\"" + filePath + "\""  + " \"" + mp4FilePath + "\"")
    count = count + 1
print ("Done the total number of file was be converted: ", count)
```

```python
# Function to extract frames
def FrameCapture(path, x):
    # Path to video file
    vidObj = cv2.VideoCapture(path)
    # Used as counter variable
    count = 0
    # checks whether frames were extracted
    success = 1

    while success:
        # vidObj object calls read
        # function extract frames
        success, image = vidObj.read()
        if success != 1:
            print("Empty file")
            break
        # Saves the frames with frame-count
        name = FRAMES + '/{x}frame{count}.jpg'.format(x= x, count=count)
        cv2.imwrite(name, image)
        count += 1
    vidObj.release()
```

```python

i = 2400

for file in sorted(os.listdir(VID_FILES)):
  if file.endswith('mp4'):
    i+=1
    filePath = os.path.join(VID_FILES, file)
    if os.path.isfile(filePath):
      # Calling the function
      FrameCapture(filePath, i)
    else:
      continue
```
